```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Specify the countries to highlight in green and their label
green_countries = {
    'Colombia': 'Implemented regulation', 'Mongolia': 'Implemented regulation',
    'China': 'Implemented regulation', 'Japan': 'Implemented regulation',
    'Malaysia': 'Implemented regulation', 'Austria': 'Implemented regulation',
    'Belgium': 'Implemented regulation', 'Bulgaria': 'Implemented regulation',
    'Croatia': 'Implemented regulation', 'Republic of Cyprus': 'Implemented regulation',
    'Czech Republic': 'Implemented regulation', 'Denmark': 'Implemented regulation',
    'Estonia': 'Implemented regulation', 'Finland': 'Implemented regulation',
    'France': 'Implemented regulation', 'Germany': 'Implemented regulation',
    'Greece': 'Implemented regulation', 'Hungary': 'Implemented regulation',
    'Ireland': 'Implemented regulation', 'Italy': 'Implemented regulation',
    'Latvia': 'Implemented regulation', 'Lithuania': 'Implemented regulation',
    'Luxembourg': 'Implemented regulation', 'Malta': 'Implemented regulation',
    'Netherlands': 'Implemented regulation', 'Poland': 'Implemented regulation',
    'Portugal': 'Implemented regulation', 'Romania': 'Implemented regulation',
    'Slovakia': 'Implemented regulation', 'Slovenia': 'Implemented regulation',
    'Spain': 'Implemented regulation', 'Sweden': 'Implemented regulation'
}

# Specify the countries to highlight in yellow and their label
yellow_countries = {
    'Canada': 'Taxonomy in development', 'Mexico': 'Taxonomy in development',
    'Dominican Republic': 'Taxonomy in development', 'Peru': 'Taxonomy in development',
    'Brazil': 'Taxonomy in development', 'Chile': 'Taxonomy in development',
    'United Kingdom': 'Taxonomy in development', 'South Africa': 'Taxonomy in development',
    'India': 'Taxonomy in development', 'Bangladesh': 'Taxonomy in development',
    'Sri Lanka': 'Taxonomy in development', 'New Zealand': 'Taxonomy in development',
    'Russia': 'Taxonomy in development', 'Kazakhstan': 'Taxonomy in development',
    'Thailand': 'Taxonomy in development', 'South Korea': 'Taxonomy in development',
    'Vietnam': 'Taxonomy in development', 'Philippines': 'Taxonomy in development',
    'Singapore': 'Taxonomy in development', 'Indonesia': 'Taxonomy in development'
}

# Filter the world dataframe to include only the specified countries
highlighted_green = world[world['name'].isin(green_countries.keys())]
highlighted_yellow = world[world['name'].isin(yellow_countries.keys())]

# Plot the world map
fig, ax = plt.subplots(figsize=(12, 8))
world.plot(color='lightgray', edgecolor='white', linewidth=0.5, ax=ax)

# Plot the highlighted countries in green
highlighted_green.plot(ax=ax, color='#2c344c')

# Plot the highlighted countries in yellow
highlighted_yellow.plot(ax=ax, color='#ea5a51')

# Remove axis labels and ticks
ax.axis('off')

# Add color legend
ax.text(0.2, -0.1, '- Implemented regulation', fontsize=12, transform=ax.transAxes)
ax.text(0.6, -0.1, '- Taxonomy in development', fontsize=12, transform=ax.transAxes)
ax.add_patch(plt.Rectangle((0.05, -0.15), 0.1, 0.05, color='#2c344c'))
ax.add_patch(plt.Rectangle((0.45, -0.15), 0.1, 0.05, color='#ea5a51'))

# Display the plot
plt.show()
```
